```python
def bboxTransform(bbox):
    bbox[2]=bbox[0]+bbox[2]
    bbox[3]=bbox[1]+bbox[3]
    return bbox
```

```python
def reformatGT(path):
    f = open('./data-airport/annotations/set01/{}'.format(path),'r')
    inputLines = []
    for lines in f:
        a = str(lines)
        #print a
        inputLines.append(a)
    f.close()
    inputLines.reverse()
    inputLines.pop()
    w = open('./data-airport/ground-truth/{}'.format(path),'w')
    for lines in inputLines:
        a = lines.split(" ")
        label = [i for i in a[0:1]]
        bbox = [float(i) for i in a[1:5]]
        bbox = bboxTransform(bbox)
        if label[0] != 'ignore' :
            line = label + [str(i) for i in bbox]
            line[4]+='\n'
            line = " ".join(line)
            w.writelines(line)
    w.close()


```

```python
for root, subdir, fnames in os.walk('./data-airport/annotations/set01'):
    for fname in fnames:
        #print os.path.join(root,fname)
        reformatGT(fname)
```
